```python
df_prod = df_produtos

for ind in range(0,3):
    try:            
        print('\n'+str(int((ind/len(df_prod))*100))+"%") #Indicando o andamento
        print('Acessesando a pagina do produto '+ str(df_prod['Produto'][ind]) +'...')
        time.sleep(tempo)
        html = urlopen(df_prod['Link'][ind])    
        bsObj = BeautifulSoup(html.read(), "lxml") 


  
        # Dados Gerais
        df_prod.loc[ind,'Coleção'] = bsObj.find('section',{'class':'container'}).find('ul',{'class':'tags'}).get_text(strip=True)


        # Tipo e Onde Comercializa
        div_info = bsObj.find('section',{'class':'container'}).find('div',{'class':'info'}).find_all('div',{'class':'compositions-content'})

        div_tipo_produto = div_info[0]
        div_ondecompra_produto = div_info[1]

        str_somatipo = ''
        for somandotipo in div_tipo_produto.find_all('a',{'class':'button'}):
            str_somatipo += somandotipo.get_text(strip=True) + '. '

        str_somaondecompra = ''
        for somandoondecompra in div_ondecompra_produto.find_all('a',{'class':'button'}):
            str_somaondecompra += somandoondecompra.get_text(strip=True) + '. '


        df_prod.loc[ind,'Coleção'] = str_somatipo
        df_prod.loc[ind,'Onde Comercializa'] = str_somaondecompra


        # Informação técnica
        aaaa__ = bsObj.find_all('section',{'class':'technical-information'})[1]
        for teste_titulo in aaaa__.find_all('li'):
            try:
                campo = teste_titulo.find('span',{'class':'title'}).get_text(strip=True)
            
                atributo_somado = '-'
                # Condicional apenas para tratar o dado caso haja mais de um atributo no campo
                if len(teste_titulo.find_all('li',{'class':'sub-item'})) != 1:            
                    for atributo in teste_titulo.find_all('li',{'class':'sub-item'}):
                        atributo_somado += atributo.get_text(strip=True) + '. '
                else:
                    atributo_somado = teste_titulo.find('li',{'class':'sub-item'}).get_text(strip=True)                 
            except:
                df_prod.loc[ind,campo] = atributo_somado 
                pass

        # Beneficios
        soma_beneficios = ''
        for beneficios in bsObj.find('section',{'class':'benefits technical-information'}).find('ul',{'class':'technical-information-list'}).find_all('li'):
            soma_beneficios += beneficios.get_text(strip=True) + '. '
        df_prod.loc[ind,'Beneficios'] = soma_beneficios

        df_prod.to_excel('Teste 2 Catalogo.xlsx')
    except:
        print('*** ERRO ***')

```
