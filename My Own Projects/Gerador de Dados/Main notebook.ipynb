```python
#Gerador de dados. Crie tabelas de dimensões como dim_cliente, dim_produto, dim_tempo e uma tabela de fatos como fato_vendas. Popule essas tabelas com dados fictícios para simular um ambiente de data warehouse. Use bibliotecas como Faker para gerar dados realistas. Por fim gere csv de cada dataframe
import pandas as pd
from faker import Faker
import random
import csv
from datetime import datetime, timedelta
fake = Faker()
# Gerar dados para dim_cliente
def gerar_dim_cliente(num_clientes):
    clientes = []
    for _ in range(num_clientes):
        cliente = {
            'cliente_id': fake.unique.random_int(min=1, max=1000),
            'nome': fake.name(),
            'email': fake.email(),
            'telefone': fake.phone_number(),
            'endereco': fake.address()
        }
        clientes.append(cliente)
    return pd.DataFrame(clientes)
# Gerar dados para dim_produto
def gerar_dim_produto(num_produtos):
    produtos = []
    for _ in range(num_produtos):
        produto = {
            'produto_id': fake.unique.random_int(min=1, max=1000),
            'nome': fake.word(),
            'categoria': random.choice(['Eletrônicos', 'Roupas', 'Alimentos', 'Móveis']),
            'preco': round(random.uniform(10.0, 500.0), 2)
        }
        produtos.append(produto)
    return pd.DataFrame(produtos)
# Gerar dados para dim_tempo
def gerar_dim_tempo(num_dias):
    tempo = []
    start_date = datetime.now() - timedelta(days=num_dias)
    for i in range(num_dias):
        data = start_date + timedelta(days=i)
        tempo.append({
            'tempo_id': i + 1,
            'data': data.strftime('%Y-%m-%d'),
            'ano': data.year,
            'mes': data.month,
            'dia': data.day
        })
    return pd.DataFrame(tempo)
# Gerar dados para fato_vendas
def gerar_fato_vendas(num_vendas, dim_cliente, dim_produto, dim_tempo):
    vendas = []
    for _ in range(num_vendas):
        venda = {
            'venda_id': fake.unique.random_int(min=1, max=10000),
            'cliente_id': random.choice(dim_cliente['cliente_id']),
            'produto_id': random.choice(dim_produto['produto_id']),
            'tempo_id': random.choice(dim_tempo['tempo_id']),
            'quantidade': random.randint(1, 10),
            'valor_total': round(random.uniform(20.0, 1000.0), 2)
        }
        vendas.append(venda)
    return pd.DataFrame(vendas)
# Gerar os dados
num_clientes = 100
num_produtos = 50
num_dias = 365
num_vendas = 500
dim_cliente = gerar_dim_cliente(num_clientes)
dim_produto = gerar_dim_produto(num_produtos)
dim_tempo = gerar_dim_tempo(num_dias)
fato_vendas = gerar_fato_vendas(num_vendas, dim_cliente, dim_produto, dim_tempo)
# Salvar os dataframes em arquivos CSV
dim_cliente.to_csv('dim_cliente.csv', index=False)
dim_produto.to_csv('dim_produto.csv', index=False)
dim_tempo.to_csv('dim_tempo.csv', index=False)
fato_vendas.to_csv('fato_vendas.csv', index=False)
print("Dados gerados e salvos em arquivos CSV com sucesso!")
```
