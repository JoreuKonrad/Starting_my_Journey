```python
#importar 'porcelanato_data.csv'
import pandas as pd
data = pd.read_csv('porcelanato_data.csv')
#mostrar as primeiras linhas do dataframe
data
```

```python
#apresentar a tabela de produto com o valor minimo, maximo e a diferença entre eles usando a coluna 'price numeric'
data.groupby('name')['price numeric'].agg(['min', 'max', lambda x: x.max() - x.min()]).rename(columns={'<lambda_0>': 'difference'}).sort_values('difference', ascending=False)
```

```python
#faça um histograma de 'price numeric' usando o matplotlib
import matplotlib.pyplot as plt
plt.hist(data['price numeric'], bins=20)
plt.xlabel('Price Numeric')
plt.ylabel('Frequency')
plt.title('Histogram of Price Numeric')
plt.show()
```
